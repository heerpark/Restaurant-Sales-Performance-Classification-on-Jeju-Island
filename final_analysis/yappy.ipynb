```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
# df_final: keyword 원핫 + 기존 컬럼 포함된 최종 df
# feature_cols: 원핫된 컬럼들
feature_cols = mlb.classes_.tolist()

# 종속변수 인코딩
label_encoder = LabelEncoder()
df_final['label'] = label_encoder.fit_transform(df_final['UE_AMT_GRP'])

# 결측치 제거 (입력 또는 label 중 하나라도 결측이면 제거)
df_final = df_final.dropna(subset=feature_cols + ['label'])

# 입력(X), 출력(y)
X = df_final[feature_cols].values.astype(np.float32)
y = df_final['label'].values.astype(np.int64)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

train_dataset = TensorDataset(torch.tensor(X_train), torch.tensor(y_train))
val_dataset = TensorDataset(torch.tensor(X_val), torch.tensor(y_val))

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

```python
import torch.nn as nn

class DeepMLPClassifier(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.3),           # Dropout 추가

            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, output_dim)  # 출력층 (softmax는 CrossEntropyLoss에서 자동 적용됨)
        )

    def forward(self, x):
        return self.net(x)
```

```python
model = DeepMLPClassifier(input_dim=X.shape[1], output_dim=6)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# GPU 사용 설정 (선택)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
EPOCHS = 100

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0.0

    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)

        optimizer.zero_grad()
        pred = model(xb)
        loss = criterion(pred, yb)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * xb.size(0)

    avg_loss = total_loss / len(train_loader.dataset)

    # 검증
    model.eval()
    correct = total = 0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            predicted = torch.argmax(out, dim=1)
            correct += (predicted == yb).sum().item()
            total += yb.size(0)

    val_acc = correct / total
    print(f"[Epoch {epoch+1:02d}] Train Loss: {avg_loss:.4f} | Val Acc: {val_acc:.4f}")
```

```python
from sklearn.metrics import classification_report

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for xb, yb in val_loader:
        xb = xb.to(device)
        outputs = model(xb)
        preds = torch.argmax(outputs, dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(yb.numpy())

print(classification_report(all_labels, all_preds, target_names=label_encoder.classes_))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# 종속 변수 인코딩
label_encoder = LabelEncoder()
df_final['label'] = label_encoder.fit_transform(df_final['UE_AMT_GRP'])

# 특성 정의
mlp_features = mlb.classes_.tolist()  # 원핫 키워드

catboost_features = [
    'MON_UE_CNT_RAT', 'TUE_UE_CNT_RAT', 'WED_UE_CNT_RAT', 'THU_UE_CNT_RAT', 'FRI_UE_CNT_RAT', 'SAT_UE_CNT_RAT', 'SUN_UE_CNT_RAT',
    'peak_day_label', 'day_type_encoded', 'weekday_ratio', 'weekend_ratio', 'day_entropy',
    'HR_5_11_UE_CNT_RAT', 'HR_12_13_UE_CNT_RAT', 'HR_14_17_UE_CNT_RAT', 'HR_18_22_UE_CNT_RAT', 'HR_23_4_UE_CNT_RAT',
    'peak_time_label', 'peak_ratio', 'peak_min_diff', 'top2_gap', 'time_type_label',
    'LOCAL_UE_CNT_RAT',
    'RC_M12_MAL_CUS_CNT_RAT', 'RC_M12_FME_CUS_CNT_RAT',
    'RC_M12_AGE_UND_20_CUS_CNT_RAT', 'RC_M12_AGE_30_CUS_CNT_RAT', 'RC_M12_AGE_40_CUS_CNT_RAT',
    'RC_M12_AGE_50_CUS_CNT_RAT', 'RC_M12_AGE_OVR_60_CUS_CNT_RAT', 'main_age_group_label',
    'age_entropy', 'young_ratio', 'old_ratio', 'age_concentration', 'expected_age',
    'MCT_TYPE_label', 'center_group_label'
]

cat_features = ['peak_day_label', 'day_type_encoded',
                'peak_time_label', 'time_type_label',
                'main_age_group_label', 'MCT_TYPE_label', 'center_group_label']

# 결측치 제거
df_final = df_final.dropna(subset=mlp_features + catboost_features + ['label'])

# train/val split (공통)
df_train, df_val = train_test_split(df_final, test_size=0.2, stratify=df_final['label'], random_state=42)
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

# MLP용 데이터
X_train_mlp = df_train[mlp_features].values.astype(np.float32)
X_val_mlp = df_val[mlp_features].values.astype(np.float32)
y_train = df_train['label'].values.astype(np.int64)
y_val = df_val['label'].values.astype(np.int64)

train_loader = DataLoader(TensorDataset(torch.tensor(X_train_mlp), torch.tensor(y_train)), batch_size=64, shuffle=True)
val_loader = DataLoader(TensorDataset(torch.tensor(X_val_mlp), torch.tensor(y_val)), batch_size=64, shuffle=False)

class DeeperMLPClassifier(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, output_dim)
        )

    def forward(self, x):
        return self.net(x)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = DeepMLPClassifier(input_dim=X_train_mlp.shape[1], output_dim=6).to(device)

# 학습
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ✅ 여기에 class weight 설정
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(device)

# ✅ 가중치 적용된 loss 함수 정의
criterion = torch.nn.CrossEntropyLoss(weight=class_weights_tensor)

EPOCHS = 200

for epoch in range(EPOCHS):
    model.train()
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        loss = criterion(model(xb), yb)
        loss.backward()
        optimizer.step()

    model.eval()
    correct = total = 0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            pred = model(xb)
            correct += (torch.argmax(pred, dim=1) == yb).sum().item()
            total += yb.size(0)
    if epoch % 50 == 0:
        print(f"[Epoch {epoch+1}] Val Acc: {correct/total:.4f}")
```

```python
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

model.eval()
y_pred = []
y_true = []

boost_class = 5       # 강조하고 싶은 클래스 index
boost_margin = 0.3   # 확률에 더해줄 값

with torch.no_grad():
    for xb, yb in val_loader:
        xb = xb.to(device)
        outputs = model(xb)

        # ✅ softmax → 확률로 변환
        probs = torch.softmax(outputs, dim=1).cpu().numpy()

        # ✅ 클래스 5에 boost
        probs[:, boost_class] += boost_margin

        # ✅ argmax
        preds = np.argmax(probs, axis=1)

        y_pred.extend(preds)
        y_true.extend(yb.numpy())

# 평가
print(f"📊 MLP Accuracy (threshold adjusted): {accuracy_score(y_true, y_pred):.4f}")
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_))
```

```python
from catboost import CatBoostClassifier

# 1. 모델 정의
cat_model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.1,
    depth=6,
    loss_function='MultiClass',
    eval_metric='MultiClass',
    early_stopping_rounds=50,
    random_seed=42,
    verbose=100
)

# 2. 학습
cat_model.fit(
    df_train[catboost_features], df_train['label'],
    eval_set=(df_val[catboost_features], df_val['label']),
    cat_features=cat_features
)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# 검증셋에 대한 예측
cat_preds = cat_model.predict(df_val[catboost_features])
cat_preds = cat_preds.astype(int)  # label이 int형이라면 안전하게 맞춰줌

# 정확도 확인
cat_acc = accuracy_score(df_val['label'], cat_preds)
print(f"📊 CatBoost Accuracy: {cat_acc:.4f}")

# 클래스별 정밀도, 재현율, F1 확인
print(classification_report(df_val['label'], cat_preds, target_names=label_encoder.classes_))
```
