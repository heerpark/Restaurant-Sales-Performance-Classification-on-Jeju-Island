```python
# 1. id별로 keyword를 모두 모으기
keyword_aggregated = review_df.groupby('id').agg({
    'keyword': lambda x: ', '.join(
        sum((k.split(',') for k in x.dropna()), [])  # ','로 나눠서 리스트로 만들고 다 합치기
    )
}).reset_index()

print(keyword_aggregated)
```

```python
print(keyword_aggregated.iloc[0])
```

```python
# 1. Nope 제거 함수
def clean_keywords(keyword_str):
    if pd.isna(keyword_str):
        return keyword_str
    keywords = [k.strip() for k in keyword_str.split(',')]
    filtered = [k for k in keywords if k != 'Nope']
    return ', '.join(filtered)

# 2. 긍정/부정 분리 함수
def split_positive_negative(keyword_str):
    if pd.isna(keyword_str):
        return pd.Series([None, None])  # [positive, negative]
    
    keywords = [k.strip() for k in keyword_str.split(',')]
    positive = []
    negative = []
    for k in keywords:
        if k.endswith('_긍정'):
            positive.append(k.replace('_긍정', '').strip())
        elif k.endswith('_부정'):
            negative.append(k.replace('_부정', '').strip())
    return pd.Series([
        ', '.join(positive) if positive else None,
        ', '.join(negative) if negative else None
    ])

# 3. 적용하기
# (1) Nope 제거
keyword_aggregated['keyword'] = keyword_aggregated['keyword'].apply(clean_keywords)

# (2) 긍정/부정 나누기
keyword_aggregated[['positive_keyword', 'negative_keyword']] = keyword_aggregated['keyword'].apply(split_positive_negative)

# 결과 확인
print(keyword_aggregated)
```

```python
# 1. 가장 많이 온 요일 (디버깅 포함)
def get_peak_day_debug(row):
    try:
        values = row.values.astype(float)  # float으로 강제 변환
        if np.all(values == 0):  # 전체가 0이면
            return '없음'
        peak_idx = np.argmax(values)  # 가장 큰 값의 인덱스
        return days[peak_idx]
    except Exception as e:
        print(f"Error in get_peak_day at index {row.name}: {e}")
        raise  # 에러 다시 던져서 멈추게
```

```python
# 2. 평일형 / 주말형 (디버깅 포함)
def get_type_label_debug(row):
    try:
        values = row.values.astype(float)
        if np.all(values == 0):
            return '없음'
        weekday_avg = values[:5].mean()  # 월화수목금
        weekend_avg = values[5:].mean()  # 토일
        if weekday_avg > weekend_avg:
            return '평일형'
        else:
            return '주말형'
    except Exception as e:
        print(f"Error in get_type_label at index {row.name}: {e}")
        raise
```
