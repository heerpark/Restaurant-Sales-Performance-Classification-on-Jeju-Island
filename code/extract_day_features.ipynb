```python
# 요일 이름 (월 ~ 일 순서)
days = ['월', '화', '수', '목', '금', '토', '일']
```

```python
# 1. 가장 많이 온 요일 (디버깅 포함)
def get_peak_day_debug(row):
    try:
        values = row.values.astype(float)  # float으로 강제 변환
        if np.all(values == 0):  # 전체가 0이면
            return '없음'
        peak_idx = np.argmax(values)  # 가장 큰 값의 인덱스
        return days[peak_idx]
    except Exception as e:
        print(f"Error in get_peak_day at index {row.name}: {e}")
        raise  # 에러 다시 던져서 멈추게
```

```python
# 2. 평일형 / 주말형 (디버깅 포함)
def get_type_label_debug(row):
    try:
        values = row.values.astype(float)
        if np.all(values == 0):
            return '없음'
        weekday_avg = values[:5].mean()  # 월화수목금
        weekend_avg = values[5:].mean()  # 토일
        if weekday_avg > weekend_avg:
            return '평일형'
        else:
            return '주말형'
    except Exception as e:
        print(f"Error in get_type_label at index {row.name}: {e}")
        raise
```

```python
# 적용 (이때 필요한 컬럼들만 apply)
# df의 MON_UE_CNT_RAT ~ SUN_UE_CNT_RAT 컬럼만 골라서 작업
cols = ['MON_UE_CNT_RAT', 'TUE_UE_CNT_RAT', 'WED_UE_CNT_RAT', 
        'THU_UE_CNT_RAT', 'FRI_UE_CNT_RAT', 'SAT_UE_CNT_RAT', 'SUN_UE_CNT_RAT']
```

```python
df['peak_day'] = df[cols].apply(get_peak_day_debug, axis=1)
```

```python
df['type_label'] = df[cols].apply(get_type_label_debug, axis=1)
```
