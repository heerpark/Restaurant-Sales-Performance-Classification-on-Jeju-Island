```python
def move_columns_after(df, move_cols, target_col):
    cols = df.columns.tolist()
    
    # 1. 옮길 컬럼 제거
    for col in move_cols:
        cols.remove(col)
    
    # 2. target_col 위치 찾기
    insert_idx = cols.index(target_col) + 1
    
    # 3. target_col 뒤에 move_cols 삽입
    new_cols = cols[:insert_idx] + move_cols + cols[insert_idx:]
    
    # 4. 재정렬된 DataFrame 반환
    return df[new_cols]

# 예시
move_cols = ['peak_day_label']
target_col = 'UE_AMT_GRP_encoded'

df = move_columns_after(df, move_cols, target_col)
```

```python
df.columns
```

```python
# 원핫인코딩
peak_day_dummies = pd.get_dummies(df['peak_day'], prefix='peak_day')

# 원래 df에 붙이기
df = pd.concat([df, peak_day_dummies], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# 1. 라벨 인코더 생성
le = LabelEncoder()

# 2. 'center_group' 컬럼 라벨 인코딩
result['center_group_label'] = le.fit_transform(result['center_group'])

# 3. 라벨 인코딩 매핑 확인
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

# 4. 'center_group' 컬럼 바로 뒤에 새 컬럼 넣기
cols = result.columns.tolist()
center_idx = cols.index('center_group')
new_order = cols[:center_idx + 1] + ['center_group_label'] + cols[center_idx + 1:]
result = result[new_order]

# 5. 매핑 출력
print(label_mapping)
```
